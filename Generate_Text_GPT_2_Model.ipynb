```python
from transformers import pipeline

model = pipeline("text-generation", model = "gpt2")
```

```python
sentence = model("Hi, My name is Madhav Mishra, I am data scientist", 
                 do_sample=True, top_k=100, 
                 temperature=0.9, max_length=300, 
                 num_return_sentences=3)
```

```python
for i in sentence:
  print(i["generated_text"])
```
